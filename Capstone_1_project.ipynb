```python
import numpy as np 

import pandas as pd 
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
import json
from geopy.geocoders import Nominatim 
import geocoder 
import requests
from bs4 import BeautifulSoup 
from pandas.io.json import json_normalize 
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium 
print("All Libraries imported!!")
```

```python
source = requests.get("https://en.wikipedia.org/wiki/Category:Districts_of_Cairo").text
soup = BeautifulSoup(source, 'html.parser')
neighborhoodL = []
```

```python
for row in soup.find_all("div", class_="mw-category")[0].findAll("li"):
    neighborhoodL.append(row.text)
```

```python
r_df = pd.DataFrame({"Neighborhood": neighborhoodL})

r_df.head()
```

```python
def get_latlng(neighborhood):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, cairo, Egypt'.format(neighborhood))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
coords = [ get_latlng(neighborhood) for neighborhood in r_df["Neighborhood"].tolist() ]
coords
```

```python
df_coords = pd.DataFrame(coords, columns=[ 'Longitude', 'Latitude'])
```

```python
r_df['Longitude'] = df_coords['Longitude']
r_df['Latitude'] = df_coords['Latitude']
```

```python
print(r_df.shape)
r_df
```

```python
r_df.to_csv("r_df.csv", index=False)
```

```python
address = 'Cairo, Egypt'

geolocator = Nominatim(user_agent="my-application")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of cairo, Egypt {}, {}.'.format(latitude, longitude))
```

```python
map_r = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, neighborhood in zip(r_df['Latitude'], r_df['Longitude'], r_df['Neighborhood']):
    label = '{}'.format(neighborhood)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_r)  
    
map_r
```

```python
radius = 2000
LIMIT = 100

venues = []

for lat, long, neighborhood in zip(r_df['Latitude'], r_df['Longitude'], r_df['Neighborhood']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    for venue in results:
        venues.append((
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
venues_df = pd.DataFrame(venues)
venues_df.columns = ['Neighborhood', 'Latitude', 'Longitude', 'VenueName', 'VenueLongitude', 'VenueLatitude', 'VenueCategory']
print(venues_df.shape)
venues_df.head()
```

```python
venues_df['VenueCategory'].unique()[:50]
```

```python
"Shopping Mall" in venues_df['VenueCategory'].unique()
```

```python
r_xyz = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")
r_xyz['Neighborhoods'] = venues_df['Neighborhood'] 
print(r_xyz.shape)
r_xyz.head()
```

```python
r_grouping = r_xyz.groupby(["Neighborhoods"]).mean().reset_index()

print(r_grouping.shape)
r_grouping.head()
```

```python
len(r_grouping[r_grouping["Shopping Mall"] > 0])
```

```python
r_complex = r_grouping[["Neighborhoods","Shopping Mall"]]
r_complex.head()
```

```python
clusters = 3

r_clustering = r_complex.drop(["Neighborhoods"], 1)
k_means = KMeans(n_clusters=clusters, random_state=0).fit(r_clustering)
k_means.labels_[0:9]

```

```python
r_join=r_complex.copy()
r_join['labels']=k_means.labels_
r_join.head()
```

```python
r_join = r_join.join(r_df.set_index("Neighborhood"), on="Neighborhoods")

print(r_join.shape)
r_join.head()
```

```python
r_join
```

```python
map_c = folium.Map(location=[latitude, longitude], zoom_start=11)
a = np.arange(clusters)
cal = [i+a+(i*a)**2 for i in range(clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(cal)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
markers_colors = []
for lat, lon, poi, cluster in zip(r_join['Latitude'], r_join['Longitude'], r_join['Neighborhoods'], r_join['labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_c)
       
map_c
```
